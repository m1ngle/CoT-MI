```python
import os
from openai import OpenAI

# The client automatically picks up the API key from the OPENAI_API_KEY environment variable
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Define the messages for the chat completion
# Role-based messages help control the model's behavior
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Explain how an API works in a simple sentence."}
]

try:
    # Make the API call to generate a response
    response = client.chat.completions.create(
        model="gpt-3.5-turbo", # Specify the model you want to use
        messages=messages,
        temperature=0.7,      # Controls randomness (0.0 to 1.0)
        max_tokens=150        # Limits the response length
    )

    # Extract and print the model's response
    print(response.choices[0].message.content.strip())

except Exception as e:
    print(f"An error occurred: {e}")
```
