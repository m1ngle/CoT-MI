```python
import json
import tiktoken
from pathlib import Path

# Define the functions
def count_words(sentence: str) -> int:
    return len(sentence.strip().split())

enc = tiktoken.get_encoding("cl100k_base")
def count_tokens(sentence: str) -> int:
    return len(enc.encode(sentence))

# Get all JSON files in the current directory
json_file = 'stimuli/ambiguous_stimuli.json'

   
    # Read the JSON file
with open(json_file, 'r') as f:
    data = json.load(f)

    for item in data:
        item['context_length'] = count_words(item['sentence'])
        item['context_length_tokens'] = count_tokens(item['sentence'])

    
    # Write back to the file
with open(json_file, 'w') as f:
        json.dump(data, f, indent=2)
    
print(f"✓ Updated {json_file}")

```

```python
# Get all JSON files in the current directory
json_files = Path('.').glob('*.json')

combined_data = []

for json_file in json_files:
    print(f"Reading {json_file}...")
    
    with open(json_file, 'r') as f:
        data = json.load(f)
    
    # Add data to combined list
    if isinstance(data, list):
        combined_data.extend(data)
    elif isinstance(data, dict):
        # If each file is a single dict, add it as an item
        combined_data.append(data)

# Write combined data to stimuli.json
with open('stimuli/stimuli.json', 'w') as f:
    json.dump(combined_data, f, indent=2)

print(f"\n✓ Combined {len(combined_data)} items into stimuli.json")
```
