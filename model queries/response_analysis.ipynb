```python
sub_wide_df = sub_wide.reset_index()
sub_wide_df.head()
```

```python
#Function to run McNemar Tests
from statsmodels.stats.contingency_tables import mcnemar

def run_mcnemar(df, colA, colB):

    tmp = df[[colA, colB]].dropna().copy()

    table = pd.crosstab(tmp[colA], tmp[colB])
    table = table.reindex(index=[False, True], columns=[False, True], fill_value=0)

    b = table.loc[True, False]
    c = table.loc[False, True]
    n = b + c
#adding different tests per number discordant pairs
    if n <= 25:
        result = mcnemar(table, exact=True)
        test_type = "exact"
    else:
        result = mcnemar(table, exact=False, correction=True)
        test_type = "asymptotic"

    return table, result.pvalue, n, test_type
```

```python
# Stratified comparison conditions
comparison_specs = {
    "CoT vs Direct": [
        ("api_cot_zero", "api_direct_zero"),
        ("api_cot_few", "api_direct_few"),
        ("manual_cot_zero", "manual_direct_zero"),
        ("manual_cot_few", "manual_direct_few"),
    ],
    "Few vs Zero": [
        ("api_cot_few", "api_cot_zero"),
        ("api_direct_few", "api_direct_zero"),
        ("manual_cot_few", "manual_cot_zero"),
        ("manual_direct_few", "manual_direct_zero"),
    ],
    "Manual vs API": [
        ("manual_cot_zero", "api_cot_zero"),
        ("manual_cot_few", "api_cot_few"),
        ("manual_direct_zero", "api_direct_zero"),
        ("manual_direct_few", "api_direct_few"),
    ],
    "CoT Few vs CoT Zero": [
        ("api_cot_few", "api_cot_zero"),
        ("manual_cot_few", "manual_cot_zero"),
    ],
    "Direct Few vs Direct Zero": [
        ("api_direct_few", "api_direct_zero"),
        ("manual_direct_few", "manual_direct_zero"),
    ],
}

results = []
```

```python
for model_name in sub_wide_df["model"].dropna().unique():
    df_model = sub_wide_df[sub_wide_df["model"] == model_name].copy()

    for comparison_name, pairs in comparison_specs.items():
        for colA, colB in pairs:
            table, pvalue = run_mcnemar(df_model, colA, colB)

            # Safely extract 2x2 counts even if some combinations are absent
            a = table.loc[True, True] if True in table.index and True in table.columns else 0
            b = table.loc[True, False] if True in table.index and False in table.columns else 0
            c = table.loc[False, True] if False in table.index and True in table.columns else 0
            d = table.loc[False, False] if False in table.index and False in table.columns else 0

            results.append({
                "model": model_name,
                "comparison": comparison_name,
                "condition_A": colA,
                "condition_B": colB,
                "both_correct": a,
                "A_correct_B_wrong": b,
                "A_wrong_B_correct": c,
                "both_wrong": d,
                "p_value": pvalue
            })

results_df = pd.DataFrame(results)
results_df.head()
```

```python
results_df.to_csv("mcnemar_results.csv",index=False)
```

```python
#Primary tests
#API vs Manual
api_manual_wide = pd.pivot_table(
    sub_first,
    values='correct',
    index=['sentence', 'model'],
    columns='query_mode',
    aggfunc='first'
).reset_index()
api_manual_wide.head()
```

```python
table, pvalue, n, test_type = run_mcnemar(api_manual_wide, 'api', 'manual')
api_man_result_df = pd.DataFrame([{
    "comparison": "API vs Manual",
    "both_correct": table.loc[True, True],
    "api_correct_manual_wrong": table.loc[True, False],
    "api_wrong_manual_correct": table.loc[False, True],
    "both_wrong": table.loc[False, False],
    "discordant_pairs": n,
    "test_type": test_type,
    "p_value": pvalue
}])
api_man_result_df
```

```python
#Primary tests
#CoT vs Direct
cot_direct_wide = pd.pivot_table(
    sub_first,
    values='correct',
    index=['sentence', 'model'],
    columns='prompt_style',
    aggfunc='first'
).reset_index()

table, pvalue, n, test_type = run_mcnemar(cot_direct_wide, 'cot', 'direct')
cot_direct_result_df = pd.DataFrame([{
    "comparison": "CoT vs Direct",
    "both_correct": table.loc[True, True],
    "api_correct_manual_wrong": table.loc[True, False],
    "api_wrong_manual_correct": table.loc[False, True],
    "both_wrong": table.loc[False, False],
    "discordant_pairs": n,
    "test_type": test_type,
    "p_value": pvalue
}])
cot_direct_result_df
```

```python
#Primary tests
#Zero vs Few
zero_few_wide = pd.pivot_table(
    sub_first,
    values='correct',
    index=['sentence', 'model'],
    columns='shot',
    aggfunc='first'
).reset_index()

table, pvalue, n, test_type = run_mcnemar(zero_few_wide, 'zero', 'few')
zero_few_wide_result_df = pd.DataFrame([{
    "comparison": "Zero vs Few",
    "both_correct": table.loc[True, True],
    "api_correct_manual_wrong": table.loc[True, False],
    "api_wrong_manual_correct": table.loc[False, True],
    "both_wrong": table.loc[False, False],
    "discordant_pairs": n,
    "test_type": test_type,
    "p_value": pvalue
}])
zero_few_wide_result_df
```
